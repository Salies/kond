```python
from demoparser2 import DemoParser
import pandas as pd
import numpy as np
```

```python
parser = DemoParser("demos/003733152704298483830_0335691126.dem")
```

```python
#parser.parse_header()
```

```python
steamid_name_map = parser.parse_player_info()[["steamid", "name"]].set_index("steamid").to_dict()["name"]
steamid_name_map
```

```python
all_events = parser.parse_events(["player_death", "cs_win_panel_match", "round_officially_ended"], other=["total_rounds_played"])
```

```python
end_match_df = next((df for event_name, df in all_events if event_name == 'cs_win_panel_match'), None)
end_match_df
```

```python
last_tick = end_match_df["tick"].max()
rounds_played = end_match_df["total_rounds_played"].max()
last_tick, rounds_played
```

```python
player_properties = ["kills_total", "deaths_total", "damage_total"]
player_data_df = parser.parse_ticks(player_properties, ticks=[last_tick])
player_data_df["kpr"] = player_data_df["kills_total"] / rounds_played
player_data_df["dpr"] = player_data_df["deaths_total"] / rounds_played
player_data_df["adr"] = player_data_df["damage_total"] / rounds_played
player_data_df
```

```python
deaths_df = next((df for event_name, df in all_events if event_name == 'player_death'), None)
deaths_df
```

```python
multikills = deaths_df[["attacker_steamid", "total_rounds_played"]].groupby(["attacker_steamid", "total_rounds_played"]).size().reset_index(name="kills")
multikills = multikills[multikills["kills"] > 1]
multikills = multikills["attacker_steamid"].value_counts().reset_index(name="multikills")
multikills["proportion"] = multikills["multikills"] / rounds_played
multikills
```

```python
round_winners = parser.parse_event("round_end")
round_winners
```

```python
openers = deaths_df.loc[deaths_df.groupby('total_rounds_played')['tick'].idxmin(), ['total_rounds_played', 'attacker_steamid', 'tick']]
openers["attacker_steamid"] = openers["attacker_steamid"].astype(int)
openers_steamids = openers["attacker_steamid"].astype(int).values
opener_ticks = openers["tick"].values
openers_teams = parser.parse_ticks(["team_name"], ticks=opener_ticks, players=openers_steamids)
openers = openers.merge(openers_teams, left_on=['tick', 'attacker_steamid'], right_on=['tick', 'steamid'], how='left')
openers = openers[["total_rounds_played", "attacker_steamid", "team_name"]]
openers["team_name"] = openers["team_name"].replace({"TERRORIST": "T"})
openers.columns = ["round", "opener_steamid", "opener_team"]
openers["winner_team"] = round_winners["winner"]
openers["conversion"] = openers["opener_team"] == openers["winner_team"]
openers
```

```python
openers_counts = pd.DataFrame(openers["opener_steamid"].value_counts()).reset_index()
openers_counts = openers_counts.join(
    pd.DataFrame(openers[openers["conversion"] == True]["opener_steamid"].value_counts()).reset_index().set_index("opener_steamid"),
    on="opener_steamid",
    rsuffix="_conv"
)
openers_counts = openers_counts.fillna(0)
openers_counts["conv_proportion"] = openers_counts["count_conv"] / openers_counts["count"]
openers_counts["round_proportion"] = openers_counts["count_conv"] / rounds_played
openers_counts
```

```python
impact_formula = lambda a, b, c: 3.73761914 * a + 2.886383 * b + 0.09009486 * c + 0.07304849
```

```python
kast_data = {}
for r in range(rounds_played):
    kast_data[r] = {}
    for p in steamid_name_map:
        kast_data[r][p] = False

current_round = 0
attacker_victim_round = {}

for index, row in deaths_df.iterrows():
    roundno = row['total_rounds_played']
    attacker_id = row['attacker_steamid']
    victim_id = row['user_steamid']
    if attacker_id is None or victim_id is None:
        continue
    attacker_id = int(attacker_id)
    victim_id = int(victim_id)
    if roundno != current_round:
        # if survived
        for p in steamid_name_map.keys():
            if p not in attacker_victim_round.values():
                kast_data[roundno][p] = True

        current_round = roundno
        attacker_victim_round = {}
    attacker_victim_round[attacker_id] = victim_id
    # if killed
    kast_data[roundno][attacker_id] = True
    assister_id = row['assister_steamid']
    # if assisted
    if assister_id:
        assister_id = int(assister_id)
        kast_data[roundno][assister_id] = True
    # if traded
    # if victim in attacker_victim_round as attacker, then victim was traded
    if victim_id in attacker_victim_round:
        # get who died to the victim
        traded_id = attacker_victim_round[victim_id]
        kast_data[roundno][traded_id] = True
```

```python
kast_df = pd.DataFrame(kast_data)
# sum each row
kast_df["kast"] = kast_df.sum(axis=1) / rounds_played
kast_df["steamid"] = kast_df.index
kast_df = kast_df.reset_index(drop=True)
kast_df = kast_df[['steamid', 'kast']]
kast_df
```

```python
# "kast", "kpr", "dpr", "impact", "adr", bias
# 0.79697728,  0.42331173, -0.39860167,  0.2394093 ,  0.00280917, -0.00879399
def rating_formula(kast, kpr, dpr, impact, adr):
    res = 0
    res += 0.79697728 * kast
    res += 0.42331173 * kpr
    res += -0.39860167 * dpr
    res += 0.2394093 * impact
    res += 0.00280917 * adr
    res += -0.00879399
    return res
```

```python
def get_rating(player_id):
    print(player_id)
    rounds_w_multikill = multikills[multikills["attacker_steamid"] == str(player_id)]["proportion"]
    if len(rounds_w_multikill) == 0:
        rounds_w_multikill = 0
    else:
        rounds_w_multikill = rounds_w_multikill.values[0]

    opening_kills_per_round = openers_counts[openers_counts["opener_steamid"] == player_id]["round_proportion"]
    if len(opening_kills_per_round) == 0:
        opening_kills_per_round = 0
    else:
        opening_kills_per_round = opening_kills_per_round.values[0]

    winpct_after_opening_kill = openers_counts[openers_counts["opener_steamid"] == player_id]["conv_proportion"]
    if len(winpct_after_opening_kill) == 0:
        winpct_after_opening_kill = 0
    else:
        winpct_after_opening_kill = winpct_after_opening_kill.values[0]

    impact = impact_formula(rounds_w_multikill, opening_kills_per_round, winpct_after_opening_kill)

    kast = kast_df[kast_df["steamid"] == player_id]["kast"].values[0]
    kpr = player_data_df[player_data_df["steamid"] == player_id]["kpr"].values[0]
    dpr = player_data_df[player_data_df["steamid"] == player_id]["dpr"].values[0]
    adr = player_data_df[player_data_df["steamid"] == player_id]["adr"].values[0]

    print(kast, kpr, dpr, impact, adr)

    return rating_formula(kast, kpr, dpr, impact, adr)
```

```python
df_rating = {}
for player_id in steamid_name_map.keys():
    player_name = steamid_name_map[player_id]
    df_rating[player_name] = np.round(get_rating(player_id), 2)
df_rating = pd.DataFrame(df_rating.items(), columns=["name", "rating"])
df_rating
```
